## Iris Veri Seti KEŞİFÇİ VERİ ANALİZİ (Exploratory Data Analysis - EDA)

İris çiçeği veri seti veya Fisher'ın İris veri seti , İngiliz istatistikçi ve biyolog Ronald Fisher tarafından 1936 tarihli "Taksonomik problemlerde çoklu ölçümlerin kullanılması , doğrusal diskriminant analizinin bir örneği olarak" adlı makalesinde kullanılan ve ünlenen çok değişkenli bir veri setidir.

Iris çiçeği veri seti (iris_excel.xlsx) 3 adet çiçek türünden oluşmaktadır ve her birinden 50 adet olmak üzere: Iris setosa,Iris Virnica ve Iris Versicolor olarak adlandırılmaktadır.

Veri setimiz çok değişkenli bir veri setidir ve her numune 4 adet özelliğe sahiptir. Bunlar;

sepal.length,sepal width,petal length ve petal width

Bu veri setimizde saydığımız sepal width,sepal length vepetal width bağımsız değşkenlerimiz ve bu değişkenlere bağımlı olarakta çiçeğimizin türü Setosa, Virginica ve Versicolor olarak belirleniyor.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("iris_excel.xlsx")
```

Veri setinin çanak yaprak ve taç yaprak özelliklerinin standart sapmalaları düşük olduğunu yani bu verilerimin normal seyrettiğini gözlemleriz.

Standart sapması en yüksek olan çanak yaprak uzunluğuna (petal.length) baktığımızda ortalamasının 3.75

minimum değeri 1 ve max değerinin 6.9 olduğunu görüyoruz.

```python
sns.set(style="ticks")
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x= 'variety',y='sepal.length', data=df)
plt.subplot(2,2,2)
sns.boxplot(x= 'variety',y='sepal.width', data=df)
plt.subplot(2,2,3)
sns.boxplot(x= 'variety',y='petal.length', data=df)
plt.subplot(2,2,4)
sns.boxplot(x= 'variety',y='petal.width', data=df)
plt.show()
```
