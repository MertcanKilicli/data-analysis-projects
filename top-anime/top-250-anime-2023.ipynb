```python
# create the correlation matrix
corr_matrix = df.corr()

# create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# show the plot
plt.show()
```

```python
# Create a scatterplot of the score by genre
sns.scatterplot(x="Score", y="Popularity", data=df)
plt.show()
```
